# Color and color spaces

In this notebook, we explore the concept of color in images by working on a test image. Using the opencv library, we read in an image, display the image, and play around with the image's color, moving from one color space to another. actually, we would only look at only the RGB and HSV color spaces.

import all required libraries

```python
import cv2
import numpy as np
```

We read in the test image. to read in an image with opencv, we use the `imread` function.

```python
test_image = cv2.imread("test_image.jpeg")
```

We've just read in the test image. Let's view it's content. (print it)

```python
print(test_image)
```

An array of numbers? yes an image is represented by an array of numbers. How do we display the image?

well, there are libraries to use. One is matplotlib. Or we can just still use opencv, but, well ...<br/>
-importing matplotlib-

```python
import matplotlib.pyplot as plt
```

Now, we can display the image using the `imshow` function

```python
plt.imshow(test_image)
```

The figure is too small. Let's make it bigger.

```python
plt.figure(figsize=(15,15))
plt.imshow(test_image)
```

There is something wrong with this image, not obvious though. OpenCV reads images by default using a color space called BGR (Blue-Green-Red), the reverse of RGB. Matplotlib always plots images using the RGB color space.

what do we do now?

Yeah, BGR to RGB!

We can convert the test image from BGR to RGB by using the `cvtColor` function of openCV, and using the constant `COLOR_BGR2RGB`

```python
test_image = cv2.cvtColor(test_image, cv2.COLOR_BGR2RGB)
```

The image has now been converted from BGR format to RGB. Let's display it and see!

```python
plt.figure(figsize=(15,15))
plt.imshow(test_image)
```

Is the difference still unclear?

Let's check out the size of the image. This is done by calling the shape attribute of the image array (an image is an array of numbers). Let's do that below;

```python
image_shape = test_image.shape
print("shape of image is:", image_shape)
```

Image has shape (1080, 1920, 3). This is interpreted as;
- height = 1080
- width = 1920
- channels = 3

The 3 channels here correspond to the RGB values.

Let's check out the value of a pixel in the image, say, at position, (500,500).

```python
position = (500,500)
pixel_value = test_image[position]
print(f"the pixel value at position {position} is {pixel_value}")
```

We got 3 values, [7,5,102], for the pixel at position (500,500). These values corresponds to the channel values (i.e R-channel, G-channel, and B-channel (RGB))
- R => 7
- G => 5
- B => 102

The B-channel's value is far more than the values of other channels. Hence that pixel would have a color close to Blue.

Let's move to the HSV space

```python
test_image = cv2.cvtColor(test_image, cv2.COLOR_RGB2HSV)
```

display the image

```python
plt.figure(figsize=(15,15))
plt.imshow(test_image)
```

Wow!

what went wrong? the truth is that images are displayed in matplotlib using the RGB format. This is what HSV looks like when displayed as RGB.
