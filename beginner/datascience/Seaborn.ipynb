###**Seaborn is a data visualization library in Python. It is based on the concept of Matplotlib and Statistical graphs.** **Visualizing data properly help you manipulate and understand the data.**


####**Benefit of Seaborn**

i. Multiple variable comparison

ii. Different Colour palletes

iii. Multiple plot grids

iv. Univariate and bivariate visualizations - compare subset of data

v. Plots linear regression automatically



###**To use any library in python, you have to install the library**
####**How to Install Seaborn**

**i. pip install seaborn**

**ii. conda install seaborn**


Visit for in-depth knowledge on Seaborn library: https://seaborn.pydata.org/

To visualize data in seaborn, install seaborn's dependencies - Numpy, Matplotlib, Pandas and Scipy. In this notebook, Scipy was not installed.

```
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

**You can set the background of the graph to Darkgrid, White and Dark.
Uncomment each line of code to see the outcome.**

```
sns.set(style = 'darkgrid')
#sns.set(style = 'white')
#sns.set(style = 'dark')
```

```
tips = sns.load_dataset('tips')
tips.tail()        # print the last first data entry 
```

```
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```
#print the aggregate of smokers in the dataset
tips['smoker'].value_counts()
```

```
#hue is use to diiferent datapoint with colour
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker')
```

```
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time')
```

```
sns.relplot(x = 'total_bill', y = 'tip', style = 'size', data = tips)
```

```
#pallete is use to tune the colour of the graph
sns.relplot(x = 'total_bill', y = 'tip', hue = 'size', data = tips, palette = 'ch:r=-0.7, l= 0.95')
```

```
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size')
```

```
#size is use to regulate the size of the legend while sizes allows you to allocate values 
sns.relplot(x = 'total_bill', y = 'tip', data = tips, size = 'size' ,sizes = (15, 150))
```

```
sns.relplot(x = 'total_bill', y = 'tip', kind = 'line', data = tips, sort = True)
```

```
sns.relplot(x = 'total_bill', y = 'tip', kind = 'line', data = tips, sort = False)
```

```
dots = sns.load_dataset('dots').query("align == 'dots'")
dots.head()
```

```
sns.relplot(x = 'time', y = 'firing_rate', data = dots, kind = 'line', hue = 'coherence', style = 'choice')
```

```
sns.relplot(x = 'time', y = 'firing_rate', hue = 'coherence', size = 'choice', style = 'choice', kind = 'line', data = dots)
```

```
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'time', data = tips)   #col indicate column literarily
```

```
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', row = 'size', data = tips)      #row indicates row literarily
```

```
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'time', row = 'size', height=3, kind = 'line', estimator = None, data = tips)
```

```
#col_wrap groups the output according to the value specified
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'size', data = tips, col_wrap=3, height=3)   
```

####**Scatter Plot**

```
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips)
```

```
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', size = 'size', style = 'time')
```

```
sns.lineplot(x = 'time', y  = 'firing_rate', style = 'choice', hue = 'coherence', data = dots, markers = True, ci = 68)
```

### **Categorical Plotting**

```
tips = sns.load_dataset('tips')
tips.head()
```

```
titanic = sns.load_dataset('titanic')
titanic.head()
```

```
sns.catplot(x = 'day', y = 'total_bill', data = tips)
```

```
sns.catplot(x = 'day', y = 'total_bill', data = tips, jitter = False)
```

```
sns.catplot(x = 'smoker', y = 'tip', data = tips, order= ['No', 'Yes'])
```

###Take note to the kind of categorical plot in the below lines of code.

```
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm', hue = 'size')
```

```
sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips, hue = 'sex')
```

```
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```
sns.catplot(x = 'color', y = 'price', kind = 'boxen', data = diamonds.sort_values('color'))
```

```
sns.catplot(x = 'day', y = 'total_bill', kind = 'boxen', data = tips, dodge = False)
```

```
sns.catplot(x = 'total_bill', y = 'day', hue = 'sex', kind = 'violin', data = tips, split = True,)
```

```
sns.catplot(x = 'sex', y = 'survived', hue = 'class', kind = 'bar', data = titanic)
```

```
sns.catplot(x = 'sex', y = 'survived', hue = 'class', kind = 'point', data = titanic)
```

```
g = sns.catplot(x = 'day', y = 'total_bill', kind = 'violin', inner = None, data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', color = 'k', size = 3, data = tips, ax = g.ax)
```
