#Matplotlib is a data visualization library in python. It is a numerical extension of Numpy. 
#### %matplotlib inline allows you to see the plot created in the notebook. "plt.show()" can also be use to view the graph but it must be declared in individual cells.


For more information on Matplotlib library: https://matplotlib.org/3.1.1/index.html


```python
import matplotlib.pyplot as plt
#%matplotlib inline  
import numpy as np     
```

```python
# linspace is a tool in Python for creating numeric sequences. It contains the (start point, end point, number of elements)
x = np.linspace (0,5,11)  
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Function method of plotting graph
plt.plot(x,y)
```

### **In Matplotlib, the colours of the graph can change.**
Green for "g", Blue for "b", Black for "k", Yellow for "y" and so on. For more information, kindly refer to the documentation link.

```python
plt.plot(x,y,'k')    
```

By default matplotlib plot graph of unbroken lines but sometimes we want to modify the graph.
To convert the line(s) to broken line(s). Let the type of line preceed the colour of the graph.

'k--' means black broken line.

```python
plt.plot(x,y,'r--')    
```

```python
plt.plot(x,y,'r-')    # To change the line to a default line 
plt.xlabel('X Axis')    # To label the x - axis
plt.ylabel('Y Axis')      # To label the y - axis
plt.title('Graph of Y against X')       # Title of the graph
```

```python
plt.subplot(1,2,1)   # (Length, Breadth, Position/Index)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(x,y,'g')
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Bigger Graph')

axes2.plot(y,x)
axes2.set_xlabel('X Label')
axes2.set_ylabel('Y Label')
axes2.set_title('Smaller Graph')
```

```python
# Plotting graph with dimensions for rows and columns
fig,axes = plt.subplots(nrows = 2, ncols = 4)
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 4)

#tight_layout separates the subplots
plt.tight_layout()
```

```python
fig,axes= plt.subplots(nrows = 1,ncols = 2)

#Below code shows how to iterate an axes
for current_ax in axes:
    current_ax.plot(x,y)   
```

#### Figure Size and Dot Per Image

```python
# Function Method
fig = plt.figure(figsize = (8,1))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
#OO Method
#dpi is known as dot per image
fig,axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8,2),dpi = 200)

axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(x,y)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

###**Using Legend**
In a conventional graph, legend can serve as the key for a graph

```python
fig = plt.figure(figsize = (8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'X Cubed')

ax.legend()
```

```python
fig = plt.figure(figsize = (8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'X Cubed')

# Having loc in the bracket specify the location of the legend. 
ax.legend(loc = 10)
```

#### **Manipulating the graph**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'Yellow')       #For colour specification
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'Purple', linewidth = '5')       # For linewidth
#ax.plot(x,y, color = 'Purple', lw = '5')         #or lw for shortform of linewidth
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'Purple', linewidth = '1', alpha = 0.5)    #alpha controls the transparency of the line
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'Purple', linewidth = '1', ls='-')      #ls mean linestyle
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'Purple', linewidth = '1', ls='-', marker = 'o', markersize = 10, markerfacecolor = 'Yellow',
       markeredgewidth = 2, markeredgecolor = 'Red')    
```

### **Control over axis appearance - Plot Range**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, color = 'Purple', linewidth = '1', ls='-')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```

### **Plot types**
There are different types of Matplot graph.
They are: Scatter Plot, Histogram, Boxplot graph, Pie chart and so on

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range (1,1000), 100)
plt.hist(data)
```

```python
data = [np.random.normal(0, std, 100) for std in range (1,4)]
plt.boxplot(data, vert= True, patch_artist = True)
```

```python
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],
label="Boys",width=.5)
plt.bar([.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],
label="Girls", color='r',width=.5)
plt.legend()
plt.xlabel('Number')
plt.ylabel('Age')
plt.title('Weight per Gender')
```

```python
x = [1,1.5,2,2.5,3,3.5,3.6]
y = [7.5,8,8.5,9,9.5,10,10.5]
 
x1=[8,8.5,9,9.5,10,10.5,11]
y1=[3,3.5,3.7,4,4.5,5,5.2]
 
plt.scatter(x,y, label='high income low saving',color='r')
plt.scatter(x1,y1,label='low income high savings',color='b')
plt.xlabel('saving*100')
plt.ylabel('income*1000')
plt.title('Scatter Plot')
plt.legend()
```

```python
days = [1,2,3,4,5]
 
sleeping =[7,8,6,11,7]
eating = [2,3,4,3,2]
working =[7,8,7,2,2]
playing = [8,5,7,8,13]
slices = [7,2,2,13]       # Indicate the percentage of the activities
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']
 
plt.pie(slices,
  labels=activities,
  colors=cols,
  startangle=90,
  shadow= True,
  explode=(0,0.1,0,0),     # removes the percentage representaion of eating 
  autopct='%1.1f%%')      # format of representation
 
plt.title('Pie Plot')
```
